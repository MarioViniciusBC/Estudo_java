# Construtores em JAVA

## Construtor, um 'método' especial

Em Java, um construtor é um bloco de código especial, semelhante a um método, que é invocado automaticamente sempre que uma nova instância, ou objeto, de uma classe é criada utilizando a palavra-chave "**new**". A principal finalidade de um construtor é inicializar o estado desse novo objeto, o que geralmente significa atribuir valores iniciais aos seus atributos antes que ele seja efetivamente utilizado pelo programa. Para que o compilador reconheça esse bloco como um construtor, existem duas regras rigorosas de sintaxe: ele deve **possuir exatamente o mesmo nome da classe em que foi declarado** e **não pode especificar nenhum tipo de retorno**, nem mesmo o tipo "**void**".

## Exemplo:

Acompanhe o desenvolvimento abaixo:

~~~java
class Pessoa {
  protected String nome;

  // Construtor da classe mãe (construtor parametrizado)
  public Pessoa(String nome) {
    this.nome = nome;
  }
}
~~~

No código acima, declaramos a **classe Pessoa** que define o construtor parametrizado **Pessoa** que recebe como parâmetro uma string nome. Quando não definimos explicitamente nenhum construtor em uma classe, o próprio compilador do Java atua nos bastidores e insere um **construtor padrão**, que não recebe parâmetros e não realiza nenhuma ação específica além de garantir a alocação de memória na pilha de execução. No entanto, o mais comum no dia a dia do desenvolvimento é a criação de *construtores parametrizados*. Esses construtores recebem argumentos em sua declaração, permitindo que o objeto já nasça com dados específicos e obrigatórios fornecidos pelo desenvolvedor no exato momento de sua instanciação, o que previne que o objeto exista em um estado inválido ou incompleto.

~~~java
class Funcionario extends Pessoa {
  private String cargo;
  private double salario;

  // Construtor 1: Recebe todos os dados necessários
  public Funcionario(String nome, String cargo, double salario) {
    super(nome); // Invoca o construtor da classe mãe (Pessoa) ANTES de tudo
    
    this.cargo = cargo;
    this.salario = salario;  
  }

  // Construtor 2: Recebe apenas o nome (Exemplo de Sobrecarga)
  public Funcionario(String nome) {
    // Usa o this() para chamar o Construtor 1 desta mesma classe.
    // Assim, quem for instanciado apenas com o nome, ganha valores padrão.
    this(nome, "Estagiário", 1500.0);
  }

  public void exibirInfo() {
    System.out.println("Nome: " + nome + " | Cargo: " + cargo + " | Salário: R$" + salario);
  }
}
~~~

Uma característica muito poderosa da linguagem é a possibilidade de aplicar a **sobrecarga**, ou "**overloading**", aos construtores. Isso significa que uma única classe pode conter múltiplos construtores simultaneamente, desde que a *assinatura* de cada um seja única, ou seja, a **quantidade**, o **tipo** ou a **ordem** dos parâmetros recebidos deve ser diferente. A sobrecarga oferece grande flexibilidade ao design do software, pois permite que um mesmo tipo de objeto seja inicializado de maneiras variadas, moldando-se perfeitamente às diferentes informações que o sistema tem disponíveis no instante da criação daquela instância.

No exemplo que construímos, a **classe Funcionario** herda da **classe Pessoa**. Isso significa que, conceitualmente, "todo funcionário é uma pessoa". Por causa disso, a linguagem Java impõe uma regra rígida: para que um objeto da classe filha (Funcionario) seja criado, a classe mãe (Pessoa) precisa ser construída e inicializada primeiro.

Se um funcionário com nome **Bruno** passando apenas a string, o java identifica que deve usar o segundo construtor. Esse construtor aciona o this(), que redireciona o "Bruno" e mais dois valores padrão ("Estagiário" e 1500.0) para o primeiro construtor fazer o trabalho pesado. Isso evita que você precise escrever a mesma lógica de atribuição duas vezes. Perceba que aqui usarmos this com parênteses(`this()`).

### E a palavra super?

O construtor da classe `Pessoa` exige que um "nome" seja passado para ele (`public Pessoa(String nome)`). Como a classe `Funcionario` não é a dona original do atributo `nome` (ele vem de herança), ela não deve tentar configurá-lo diretamente. Em vez disso, o `Funcionario` pega o nome que recebeu e o repassa para a classe `Pessoa` fazer o seu trabalho. O `super(nome)` é exatamente isso: um telefonema para a classe mãe dizendo "Ei, inicialize a sua parte desse objeto usando este nome".

Ao  observarmos o código, veremos que o `super(nome)` é a primeiríssima instrução dentro do construtor do `Funcionario`. Isso não é uma escolha de estilo, é uma **obrigação da linguagem**. Você não pode executar nenhuma lógica, fazer contas ou atribuir valores no construtor da classe filha antes de chamar o construtor da classe mãe. A fundação (mãe) precisa existir antes do telhado (filha).

Se não for escrito o `super()` explicitamente em um construtor de uma classe filha, o compilador do Java, sorrateiramente, insere um `super()` vazio (sem parâmetros) na primeira linha por conta própria.

No código acima, se não tivesse sido colocado o `super(nome)`, o compilador tentaria colocar um `super()` vazio. Porém, a classe `Pessoa` não possui um construtor sem parâmetros (nós só criamos um que exige a String nome). O resultado? O código não iria compilar e o Java daria um erro dizendo que não encontrou um construtor padrão na classe `Pessoa`.
