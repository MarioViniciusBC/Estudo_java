# **Questão 64 - Concurso do Banco do Brasil**

Sejam as seguintes classes Java, que ocupam arquivos separados:

~~~java
public class CAx {
    protected int a;
    protected int b;

    public CAx() {
        a*=2;
        b*=3;
    }

    {
        a=1;
        b=2;
    }

    public int op1(int x) {
        return op2(x)+op3(x)+b;
    }

    public int op2(int x) {
        return x+a;
    }

    public static int op3(int x) {
        return x*2;
~~~

~~~java
public class CBy extends CAx {
    protected int a;

    public CBy() {
        a+=3;
        b+=3;
    }

    public int op2(int x) {
        return x-a;
    }

    public static int op3(int x) {
        return x*3;
    }
}
~~~

~~~java
public class Main {
    public static void main(String[] args) {
        CAx o = new CBy();
        System.out.println(o.op1(2));
    }
}
~~~

O que será exibido no console quando o método main for executado?
* (A) 10
* (B) 12
* (C) 14
* (D) 18
* (E) 20


### **Resposta**:
Para chegar a esse resultado, precisamos analisar passo a passo a execução do código, começando pela linha `CAx o = new CBy();` dentro do método **main**. O primeiro passo é entender o estado do objeto criado. Quando o construtor de **CBy** (classe filha) é invocado, o Java garante que o construtor da superclasse (**CAx**) seja executado primeiro. Dentro da classe **CAx**, existe um bloco de inicialização de instância (o trecho entre chaves `{ a=1; b=2; }` fora de qualquer método). Este bloco é executado antes do corpo do construtor de **CAx**. Portanto, inicialmente **a** vale **1** e **b** vale **2**. Em seguida, o construtor de **CAx** roda (`a*=2; b*=3;`), atualizando as variáveis de **CAx** para **a=2** e **b=6**.

Após a conclusão da parte da superclasse, a execução volta para a classe **CBy** (classe filha). Um ponto crucial aqui é que a classe **CBy** declara sua própria variável `protected int a;`. Isso causa o efeito de **ocultação (shadowing)**: a variável **a** de **CBy** é distinta da variável **a** de **CAx**. Já a variável **b** é herdada, ou seja, é a mesma referência. Quando o construtor de **CBy** é executado (**a+=3; b+=3;**), ele opera sobre o **a** da própria classe **CBy** (que inicia com **0**, resultando em **3**) e sobre o **b** herdado (que estava em **6**, resultando em **9**). O estado final do objeto é: **a (versão **CBy**) = 3, a (versão CAx, oculta) = 2, e b = 9**.

O segundo passo é a chamada do método: `o.op1(2)`. A variável **o** é do tipo **CAx**, mas aponta para uma instância de **CBy**. O método **op1** está definido apenas em **CAx**, então ele é executado lá. O código de **op1** é return `op2(x) + op3(x) + b;`. Aqui temos três partes para resolver. Primeiro, a variável **b**: como estamos dentro da classe **CAx**, ele acessa o campo **b** visível para essa instância. Como **b** foi modificado pelo construtor da subclasse (pois é herdado), seu valor é **9**.

A segunda parte é a chamada `op2(x)`. O método **op2** é um método de instância (não estático) e foi sobrescrito na classe **CBy**. Em Java, métodos de instância utilizam vinculação dinâmica (polimorfismo). Mesmo que a chamada esteja escrita dentro da classe **CAx**, a Máquina Virtual Java (**JVM**) verifica o tipo real do objeto em memória (que é **CBy**) e executa o método sobrescrito. O método **op2** em **CBy** executa `return x - a;`. Como este método está em **CBy**, ele usa o a de **CBy** (que vale **3**). O cálculo é **2 - 3**, resultando em **-1**.


A terceira parte é a chamada `op3(x)`. O método **op3** é declarado como static. Métodos estáticos utilizam vinculação estática, ou seja, são resolvidos em tempo de compilação com base no tipo da referência ou na classe onde o código está escrito. Como a chamada `op3(x)` é feita dentro da classe **CAx**, o compilador vincula isso ao método estático da própria classe **CAx**, ignorando a versão existente em **CBy** (métodos estáticos não sofrem sobrescrita polimórfica, eles são apenas ocultados). O método **op3** de **CAx** executa` return x * 2;`. O cálculo é 2 * 2 **negrito**, resultando em **4**.

Finalmente, somamos os resultados das três partes da expressão em **op1**: `op2(x)` (que retornou **-1**) + `op3(x)` (que retornou **4**) + `b` (que vale **9**). A conta final é **-1 + 4 + 9**, o que totaliza **12**. Por isso, a alternativa correta é a letra **B**. As outras alternativas estão incorretas pois falham em considerar algum desses aspectos, como o uso do a errado no shadowing, a confusão entre vinculação estática e dinâmica no método **op3**, ou o cálculo incorreto da herança da variável b.


#### **Alguns apontamentos**

1. Se o método **op1** estivesse sobrescrito (overridden) na classe **CBy**, a versão que seria executada seria a da classe **CBy**, e não a da classe **CAx**. Perceba que o método **op1** só está definido em **CAx** na questão acima. Como definimos a variável **o** como do tipo da superclasse (**CAx**), mas o objeto em si é uma instância da classe filha (**CBy**) e chamamos o método **op1** através dessa referência **o**, métodos de instância (não estáticos) são sempre vinculados ao objeto real que está na memória e não ao tipo da referência. Quando chamamos `o.op1(2)`, a **JVM** verifica o objeto apontado pela variável **o**, que é do tipo **CBy** (classe filha). Ela procura o método **op1** nessa classe. Como não há sobrescrita, a **JVM** "sobe" para a classe mãe **CAx** e executa o código de lá. No entanto, e isso é crucial, o contexto de execução (o **this**) continua sendo o objeto **CBy**. Isso é interessante porque, como o método op1 só está definido na classe mãe e o chamamos com o.op1(2), podemos ser levados a pensar que essa chamada de método está relacionada com a referência e não com a instância em memória.
    * Quando digitamos `o.op1(2)`, o compilador do Java olha estritamente para o tipo da referência (**CAx**). Ele se faz apenas uma pergunta: "Existe um contrato na classe **CAx** que garanta que o método **op1** existe?". Como a resposta é sim, o código compila. É essa checagem de segurança inicial que nos leva a pensar, equivocadamente, que a execução do método está amarrada à referência.

    * Quando rodamos programa, a Máquina Virtual Java assume o controle e, para métodos de instância, ela ignora o tipo da variável. Ela olha diretamente para a instância real na memória (**CBy**). Ela usa uma estrutura interna de mapeamento (conhecida como Virtual Method Table ou vtable) para procurar a implementação do método começando de baixo para cima. Ela bate na porta de **CBy**, procura por **op1**, não encontra e, através da herança, "sobe" para buscar o código em **CAx**. Ao fazer isso, ela leva a identidade do objeto real junto (o contexto **this**), garantindo que qualquer outra chamada feita ali dentro (como o **op2**) volte a procurar na classe **CBy**.

2. **(O contexto this):** Na chamada do método **op1** o contexto de execução (o **this**) continua sendo o objeto **CBy**. É justamente por causa disso que, quando o código dentro de `CAx.op1` chama `op2(x)`, o Java volta a olhar para a classe **CBy**. Como o **this** é do tipo **CBy**, a chamada `this.op2(x)` (o **this** é implícito) engatilha a versão sobrescrita que está na classe filha.

3. A confusão costuma surgir no que acontece dentro desse método herdado. Quando o código de **op1** (que está escrito na classe **CAx**) chama o método **op2**, ele está implicitamente chamando `this.op2(x)`. Como o **this** é o objeto **CBy** que criamos na memória, o polimorfismo entra em ação novamente. O Java verifica se **CBy** tem sua própria versão de **op2**. Como tem, essa versão do filho é a escolhida. É como se a classe mãe dissesse: "Eu forneço o algoritmo geral em **op1**, mas quando chego no passo **op2**, deixo o objeto decidir como se comportar". Por isso, mesmo estando dentro de um bloco de código escrito na classe pai, a execução "salta" para o método do filho.

4. A única exceção nessa dinâmica é o método estático **op3**, e é aqui que a "referência" ou, mais especificamente, o local da escrita do código importa. Métodos estáticos usam vinculação estática (feita na compilação). Como o método **op1** reside na classe **CAx**, quando o compilador vê a chamada **op3** lá dentro, ele "chumba" essa chamada diretamente à classe **CAx**. Ele ignora completamente o fato de o objeto ser **CBy** ou de haver outro método **op3** no filho, pois métodos estáticos não suportam polimorfismo. Portanto, a resposta para sua dúvida é que métodos comuns seguem o objeto (polimorfismo), enquanto métodos estáticos seguem a classe onde a chamada foi escrita.
