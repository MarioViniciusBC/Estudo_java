## 1. Quando usar o **super** em métodos e suas diferenças com o uso em construtores

Um cenário comum de uso da palavra **super** é na sobrescrita de métodos (overriding).

~~~java
// CLASSE PAI
class Funcionario {
    public void trabalhar() {
        System.out.println("Batendo o ponto e realizando tarefas gerais...");
    }
}
~~~

~~~java
// CLASSE FILHA
class Gerente extends Funcionario {
    
    // Sobrescrevendo o método trabalhar da classe pai
    @Override
    public void trabalhar() {
        // 1. Chama o comportamento original da classe Funcionario
        super.trabalhar();
        
        // 2. Adiciona o comportamento específico do Gerente
        System.out.println("Aprovando orçamentos e gerenciando a equipe!");
    }
}
~~~

Se for criado um objeto do tipo **Gerente** e chamar o método `trabalhar()`, o programa vai imprimir:
1. "Batendo o ponto e realizando tarefas gerais..." (veio do `super.trabalhar()`)

2. "Aprovando orçamentos e gerenciando a equipe!" (veio do código específico do gerente)

Não podemos usar `super()` sozinho dentro de um método comum tal como fazemos em construtores.
* **`super()` (com parênteses logo após a palavra):** É de uso exclusivo de construtores. Serve apenas para chamar o construtor da classe pai. Se você tentar colocar um `super()`  solto dentro de um método comum, o seu programa vai dar um erro de sintaxe.
* **`super.nomeDoMetodo()`:** É a forma correta de usar dentro de métodos. O super aqui funciona como uma referência à classe pai, permitindo que você acesse os comportamentos (métodos) dela.

~~~java
class Filha extends Pai {
    
    // 1. CONSTRUTOR: Aqui o super() é permitido e obrigatório!
    public Filha() {
        super(); // CORRETO: Chama o construtor do Pai para inicializar a classe.
    }

    // 2. MÉTODO COMUM: Aqui a regra muda.
    public void realizarTarefa() {
        
        super(); // ERRO! A linguagem não permite chamar construtor dentro de um método.
        
        super.realizarTarefa(); // CORRETO! Chama o método da classe pai.
    }
}
~~~

Sempre que você cria um objeto de uma classe filha (ex: **Gerente**), a primeira coisa que o computador precisa fazer é construir a parte que ela herda da classe mãe (ex: **Funcionario**). O `super()` é o mecanismo que dispara essa construção. A regra é clara: a classe mãe tem que nascer antes da classe filha.

Se você escrever um construtor na classe filha e não colocar o `super()` na primeira linha, o compilador vai agir por conta própria. Ele vai, de forma invisível, inserir um `super()` (sem parâmetros) bem no início do seu código.

Como o compilador sempre tenta colocar um `super()` vazio (sem parâmetros), isso só vai funcionar se a classe mãe tiver um construtor vazio.

Se a classe mãe exigir algum parâmetro para nascer (por exemplo, um nome), o compilador vai tentar colocar o `super()` vazio, não vai encontrar um construtor vazio na mãe, e o seu código vai dar erro de compilação.
